## Part 1: Matplotlib

```python
import matplotlib.pyplot as plt #  Python 2D/3D plotting library

import numpy as np # NumPy is the fundamental package for scientific computing
```

```python
def make_xy( start, stop, num ):
    x = np.linspace(start,stop, num) # x is a linear sequence of numbers
    y = np.sin(x) # y is a function of x - y(x)
    return x,y
```

```python
'''
quick plot
'''

my_x, my_y = make_xy(0,100,1000)

plt.plot(my_x,my_y)
plt.show() # This is NOT necessary in Jupyter Notebook, but it is needed if you run it in another environment.
```

```python
plt.plot(my_x,my_y,'-ob') # we added blue color and circle markers.
```

```python
fig, ax = plt.subplots( figsize=(15,8) ) #figure size is 10x6 (width x height in inches)

ax.plot(my_x, my_y,
       lw=2, # linewidth
       linestyle='--', #dashed line 
       color='red',
       marker='s', # square markers
       markerfacecolor='blue',
       markeredgecolor='yellow' 
       )


ax.grid() # add grid

xmin, xmax = 0, 50 # define x minimum and maximum
ymin, ymax = -1, 1 # define y minimum and maximum
ax.set_xlim(xmin, xmax) # set minimum and maximum 
ax.set_ylim(ymin, ymax) # set minimum and maximum

ax.set_ylabel('y = sin(x)', fontsize=20) # add label to y axis
ax.set_xlabel('x', fontsize=20) # add label to x axis

ax.set_yticks(np.linspace( ymin, ymax, 5)) # define the number of ticks (and ticklabels)
ax.set_xticks(np.linspace( xmin, xmax, 11)) # define the number of ticks (and ticklabels)
ax.tick_params(labelsize=20) # set tick labels

ax.set_title('I am learning how create figures with Python', fontsize=20)
```
